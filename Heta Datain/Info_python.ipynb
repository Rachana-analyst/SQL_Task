```python
# Importing libraries
import numpy as np
import pandas as pd
```

```python
# Data dataframe:
data = pd.read_csv("data.csv")
data
```

```python
# Subject dataframe:
subject = pd.read_csv("subject.csv")
subject
```

```python
# Modifying value 2 columns data by its square
data['value2'] = data['value2']**2
data
```

```python
# Writing into the file:
data.to_csv("data.csv", index=False)
print(data)
```

# Showcasing relationship between data.csv and subject.csv file:

```python
# Adding prefix to all columns for identifying easily:
data = data.add_prefix('data_')
data
```

```python
subject = subject.add_prefix('subject_')
subject
```

```python
# left join:
result = data.merge(subject, left_on = 'data_subject_id', right_on = 'subject_Subject_id', how = 'left')
result
```

```python
# Writing result into separate csv file:
result.to_csv("result.csv", index=False)
print(result)
```
